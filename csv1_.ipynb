```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10, 8]
act = pd.read_csv('OTg6QzA_activities.csv')
act.shape
```

```python
act.head(5)
```

```python
X=act.X.to_numpy()
Y=act.Y.to_numpy()
```

```python
X
```

```python
Y
```

```python
plt.style.use('grayscale')
plt.figure(facecolor= "white")

plt.scatter(X,Y)
plt.show()
```

```python
plt.figure(facecolor='white')
plt.hist2d(X,Y,bins=(25,25),cmap = plt.cm.Greys)
cb = plt.colorbar()
cb.set_label('frequency of appearance, bin = 25')
```

```python
plt.figure(facecolor='white')
plt.hist2d(X,Y,bins=(40,40),cmap = plt.cm.Greys)
cb = plt.colorbar()
cb.set_label('frequency of appearance, bin = 40')
```

```python
plt.figure(facecolor='white')
plt.hexbin(X,Y,gridsize= 20 ,cmap = plt.cm.Greys)
cb = plt.colorbar()
cb.set_label('frequency of appearance, gridsize = 20')
```

```python
x1 = X[:16570]
x2 = X[16570:16570*2]
x3 = X[16570*2:16570*3]
x4 = X[16570*3:]

y1 = Y[:16570]
y2 = Y[16570:16570*2]
y3 = Y[16570*2:16570*3]
y4 = Y[16570*3:]


```

```python
plt.figure(facecolor='white')

plt.subplot(2,2,1)
plt.hist2d(x1,y1,bins=20,cmap=plt.cm.Greys)
plt.title('1-Q')
plt.colorbar()

plt.subplot(2,2,2)
plt.hist2d(x2,y2,bins=20,cmap=plt.cm.Greys)
plt.title('2-Q')
plt.colorbar()

plt.subplot(2,2,3)
plt.hist2d(x3,y3,bins=20,cmap=plt.cm.Greys)
plt.title('3-Q')
plt.colorbar()

plt.subplot(2,2,4)
plt.hist2d(x4,y4,bins=20,cmap=plt.cm.Greys)
plt.title('4-Q')
plt.colorbar()

plt.show()
```
